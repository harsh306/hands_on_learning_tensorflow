### Constants , Sequences and Random Values

```python
import tensorflow as tf
import numpy as np
```

```python
a = tf.constant(0.0, shape=[2,3] )
```

```python
b = tf.constant(np.arange(6), shape=[2,3], dtype=tf.float32)
```

```python
c = tf.lin_space(start=5.0, stop=10.0, num=50, name="lin")
```

```python
d = tf.range(start=10, limit=100, delta=2, name="arange")
```

```python
with tf.Session() as sess:
    print("#################")
    print(sess.run(a))
    print("#################")
    print(sess.run(b))
    print("#################")
    print(sess.run(c))
    print("#################")
    print(sess.run(d))
```

## RANDOM TENSORS

```python
r = tf.random_normal( shape=[3,3], mean=0, stddev=2)
```

```python
r_truncated = tf.truncated_normal(shape=[3,3], mean=0, stddev=2)
```

```python
uni = tf.random_uniform(shape=[3,3], minval=0, maxval=2)
```

```python
x = tf.constant(np.arange(6), shape=[3,2] )
x_shuffled = tf.random_shuffle(value=x)
```

```python
with tf.Session() as sess:
    print("#################")
    print(sess.run(r))
    print("#################")
    print(sess.run(r_truncated))
    print("#################")
    print(sess.run(uni))
    print("#################")
    print(sess.run(x))
    print(sess.run(x_shuffled))
```

```python
X = tf.random_normal(mean=0, stddev=3, shape=[10,10,3])
x_crop = tf.random_crop(X, size=(5,5,3) )
```

```python
with tf.Session() as sess:
    print("#################")
    print(sess.run(X))
    print("#################")
    print(sess.run(x_crop))
```
